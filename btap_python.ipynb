# Bài 8:

```python
num = [1,2,3,1,2,1,1,4]
dict = {}

#tạo dict với key là số, value là số lần xuất hiện số đấy.
for ele in num:
    count = num.count(ele)
    dict.update({ele:count})
print(dict)

#in ra số mà có lần xuất hiện nhiều nhất và ít nhất trong dãy
a = []
b = []
for key, value in dict.items():
    if value == min(dict.values()):
        a.append(key)
    elif value == max(dict.values()):
        b.append(key)
print('Nhieu nhat:',', '.join(str(x) for x in b),'Be nhat:',', '.join(str(x) for x in a))

#in ra giá trị trung bình value của dict
print("Gia tri trung binh value cua dict:", sum(dict.values())/len(dict))
```

# Bài 10:

```python
dict = {}
def product(x,y):
    global dict
    
    if x not in dict.keys():
        dict.update({x:y})
    elif x in dict.keys():
        for key,value in dict.items():
            if key == x:
                value += y
                break
        dict.update({x:y})

product('banh',10)
product('keo',12)
product('dua hau',20)
print(dict)
```

```python
dict = ['o','e','a','i','u','h','w','x','y']
list_word = ['play','lie','stop','do']
for word in list_word:
    user = word
    if user == "end":
        break

    if (user[-2] == 'i') and (user[-1] == 'e'):
        user = user[0:-2] + 'ying' 
    elif (user[-2] == 'e') and (user[-1] == 'e'):
        user = user + 'ing'
    elif user[-1] == 'e':
        user = user[0:-1] + 'ing'
    elif (user[-2] in dict) and (user[-1] in dict):
        user = user + 'ing' 
    elif (user[-2] in dict):
        user = user + user[-1] + 'ing'
    else:
        user = user + 'ing'

    print('Output:',user)
```
